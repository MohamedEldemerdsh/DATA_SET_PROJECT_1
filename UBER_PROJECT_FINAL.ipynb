```python
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'], errors='coerce')
df['END_DATE*'] = pd.to_datetime(df['END_DATE*'], errors='coerce')
#######################################################################
df['DAY']=df['START_DATE*'].dt.weekday
day_label={0:'Mon',1:'Tues',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
df['DAY']=df['DAY'].map(day_label)
df
```

```python
sns.barplot(x=df['DAY'].value_counts().index,y=df['DAY'].value_counts(),data=df,palette='viridis')
plt.xlabel('DAY')
plt.ylabel('COUNT')
```

```python
df['MONTHS']=pd.DatetimeIndex(df['START_DATE*']).month
month_label = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'April', 5.0: 'May', 6.0: 'June', 7.0: 'July', 8.0: 'Aug', 9.0: 'Sep',
      10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec' }
df['MONTHS']=df['MONTHS'].map(month_label)      
```

```python
sns.barplot(x=df['MONTHS'].value_counts().index,y=df['MONTHS'].value_counts(),data=df,palette='viridis')
plt.xlabel('Months')
plt.ylabel('count')
```

```python
fig,axes=plt.subplots(2,3,figsize=(25,15))
sns.countplot(y=b1,data=df,palette='viridis',ax=axes[0,0]).set(title="START @'Cary'")
sns.countplot(y=b2,data=df,palette='viridis',ax=axes[0,1]).set(title="START @'Durham'")
sns.countplot(y=b3,data=df,palette='viridis',ax=axes[0,2]).set(title="START @'Unknown Location'")
sns.countplot(y=b4,data=df,palette='viridis',ax=axes[1,0]).set(title="START @'Morrisvile'")
sns.countplot(y=b5,data=df,palette='viridis',ax=axes[1,1]).set(title="START @'Whitebridge'")
sns.countplot(y=b6,data=df,palette='viridis',ax=axes[1,2]).set(title="START @'Islamabad'")
```
