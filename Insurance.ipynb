```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('insurance-data.csv')
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
print(df.duplicated().value_counts())
df[df.duplicated()]
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,12))
sns.scatterplot(x='bmi',y='charges',data=df,ax=axes[0,0])
sns.scatterplot(x='charges',y='age',data=df,ax=axes[0,1])
sns.scatterplot(x='bmi',y='charges',hue='sex',data=df,ax=axes[1,0])
sns.scatterplot(x='charges',y='age',hue='sex',data=df,ax=axes[1,1])
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.boxplot(x='charges',data=df,color='red',ax=axes[0]).set(title='Charge')
sns.boxplot(x='bmi',data=df,color='blue',ax=axes[1]).set(title='bmi')
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,12))
sns.countplot(x='sex',data=df,ax=axes[0,0]).set(title='Gender')
sns.countplot(x='sex',data=df,hue='smoker',ax=axes[0,1]).set(title='Gender')
sns.countplot(x='region',data=df,ax=axes[1,0]).set(title='Region')
sns.countplot(x='region',data=df,hue='sex',ax=axes[1,1]).set(title='Region')
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,12))
sns.barplot(x='sex',y='charges',data=df,ax=axes[0,0]).set(title='Gender vs Charge')
sns.barplot(x='sex',y='charges',hue='children',data=df,ax=axes[0,1]).set(title='Gender vs Charge')
sns.barplot(x='region',y='charges',data=df,ax=axes[1,0]).set(title='region vs charge')
sns.barplot(x='smoker',y='charges',hue='sex',data=df,ax=axes[1,1]).set(title='Smoker vs non-Smoker')
```
